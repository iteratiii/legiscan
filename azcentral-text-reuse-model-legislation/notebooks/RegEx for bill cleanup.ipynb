```python
import re
```

```python
text = open("allbillsregex2.txt").read()
```

```python
pattern = "[A-z][a-z]+\d+"
matches = re.findall(pattern,text)
matches[4000:4100]
```

```python
for match in matches:
    if(match!=r"hb\d+" and match!=r"Assembly\d+"):
        replacement = re.sub(r"\d+","",match)
        re.sub(match,replacement,text)
```

```python
with open('allbillsregex.txt', 'w') as f:
    f.write(text)
    f.close
```
